## DataFrames

- Dataframes são um conjunto de colunas do tipo Series
- São bidimensionais, tem o tamanho mutavel
- Documentação: https://pandas.pydata.org/docs/reference/frame.html

```python
# Criando um DataFrame
# Criar data
data = np.random.rand(20,5)
data
# Criar conteudo colunas
conteudo = list(st.ascii_uppercase)
conteudo
# Definir colunas
colunas = conteudo[:5]
# Criar dataframe
df = pd.DataFrame(data, columns=colunas)
df
```

## Tipos de variaveis no pandas e obtendo informações sobre o dataset

### Tipos de variaveis
- object - Podem ser strings ou um mix de strings
- int64 - Numeros inteiros
- float64 - Numeros flutuantes (decimais)
- bool - True or False
- datetime64 - Valores de Data e tempo( Vem da biblioteca DateTime)
- timedelta[ms] - É a diferença entre duas variaveis do tipo DateTime
- category - Variavel categorica

## Filtrando e localizando informações em um DataFrame

```python
# Exemplo de filtro

# FILTRO SIMPLES
filtro1 = df2['sex'] == 'FEMALE'
# FILTRO CONDICIONADO OU
filtro2 = (df2.sex == 'FEMALE')|(df2.species == 'Gentoo')
# FILTRO CONDICIONADO AND
filtro3 = (df2['sex']== 'FEMALE')&(df2.species == 'Gentoo')

```

```python
# filtro simples
df2[filtro1]
```

```python
# filtro condicionado ou
df2[filtro2]
```

```python
# filtro and
df2[filtro3]
```

### Localizando com o loc

- Documentação: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html

```python
# localizando com o loc -> LINHA E COLUNA
# SEM CONDICIONAL
df2.loc[5:10,['species','sex']]
```

```python
# COM CONDICIONAL
df2.loc[filtro1,:]
```

### Localizando com o iloc

- Documentação: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html

```python
# localizando com o iloc
df2.iloc[:15,:5:2]
```

## Tabelas cruzadas e medidas de frequência

```python
# Medindo frequencias com value counts
df2.species.value_counts()
df2.species.value_counts(normalize=True)
```

```python
# Criando tabelas cruzadas com crosstab
pd.crosstab(df2.species,df2.sex)
```

```python
# Media
df2.mean
df2.body_mass_g.mean
```

```python
# Quartil
# .5
df.quantile()
```
