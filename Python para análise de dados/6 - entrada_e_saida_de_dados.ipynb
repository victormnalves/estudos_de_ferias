### Arquivos-texto em partes

```python
pd.options.display.max_rows = 10 #alteramos a visualização para até 10 linhas

result = pd.read_csv('exemplos/ex6.csv')
result
pd.read_csv('exemplos/ex6.csv', nrows=5) #limitamos a visualização

chunker = pd.read_csv('exemplos/ex6.csv', chunksize=1000) #limitamos a atribuição a até 1000 linhas
#chunker não printa os seus valores, apenas armazena-os na memoria

tot = pd.Series([], dtype='float64') 
for piece in chunker:
    tot = tot.add(piece['key'].value_counts(), fill_value = 0) #passasmos os valors de chunker para a nova variavel
tot = tot.sort_values(ascending=False) #ordenamos os valores
tot[:10] #printamos os valores
    
```

### Formatos delimitados

```python
f = open('exemplos/ex7.csv') #importando um dado malformatado
reader = csv.reader(f)

for line in reader:
    print(line)

#manipulando os dados
with open('exemplos/ex7.csv') as f: #leitura do arquivo como lista
    lines = list(csv.reader(f)) #criando uma linha de listas

#separando linha de cabeçalho e linha de dados
header, values = lines[0], lines[1:]
header, values

#criando a coluna de dados com dict comprehension e zip
data_dict = {h: v for h, v in zip(header, zip(*values))}
data_dict

#criando um novo delimitador / final de linha
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL

#passando a leitura dos dados após manipularmos delimitadores
reader = csv.reader(f, dialect=my_dialect)
reader

#podemos passar delimitadores manualmete
reader = csv.reader(f, delimiter = '|')
reader
```
