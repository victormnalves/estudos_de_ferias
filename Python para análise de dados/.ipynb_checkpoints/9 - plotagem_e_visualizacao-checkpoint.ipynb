### Figuras e subplotagens

```python
# as plotagens de matplotlib ficam armazenadas em um objeto figure
fig = plt.figure()

# criando 3 subplots 2x2 na figura fig
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
# as plotagens de matplotlib ficam armazenadas em um objeto figure
fig = plt.figure()

# criando 3 subplots 2x2 na figura fig
ax1 = fig.add_subplot(2,2,1)
# ao passarmos o plot após os subplot, ele inserirá os valores ao último plot chadado
plt.plot(np.random.randn(70).cumsum(), 'k--')
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
# ao passarmos o plot após os subplot, ele inserirá os valores ao último plot chamado
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
# as plotagens de matplotlib ficam armazenadas em um objeto figure
fig = plt.figure()

# criando 3 subplots 2x2 na figura fig
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')

# podemos passar o que queremos plotar sem que seja após a atribuição
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
```

```python
# criando subplots de maneira mais eficiente / um subplot 2 x 3
fig, axes = plt.subplots(2,3) 
axes

# inserindo plots nos subplots
axes[0,0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[0,1].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
axes[0,2].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1,0].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
axes[1,1].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1,2].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
```

### Ajustando o espaçamento em torno das subplotagens

```python
# criando subplots
fig, axes = plt.subplots(2,3) 

axes[0,0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[0,1].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
axes[0,2].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1,0].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
axes[1,1].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1,2].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))

# ajustando o espaço com subplots_adjust
fig.subplots_adjust(left = None, bottom = None, right = None, top = None, wspace = None, hspace = None )
```

```python
# criando subplots
fig, axes = plt.subplots(2,3) 

axes[0,0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[0,1].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
axes[0,2].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1,0].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))
axes[1,1].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1,2].scatter(np.arange(30), np.arange(30) + 3 * np.random.rand(30))

# ajustando o espamento para zero
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
data = np.random.randn(30).cumsum()
data

plt.plot(data, 'k--',label='Default')
plt.plot(data, 'k-', drawstyle='steps-post',label='steps-post')
plt.legend(loc='best')
```

### Título, tiques, rótulos de tiques e rótulos de eixos 

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize = 'small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

### Acrescentando legendas

```python
fig = plt.figure(); ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')

# criando a legenda automáticamente
ax.legend(loc='best')
```

### Anotações e desenhos em uma subplotagem

```python
# importando o módulo datetime
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = pd.read_csv('exemplos/spx.csv', index_col=0, parse_dates=True)
data

spx = data['SPX']

spx.plot(ax=ax, style='k-')

# marcadores
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Sterns Fail'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+75),
               xytext = (date, spx.asof(date)+225),
               arrowprops = dict(facecolor = 'black', headwidth = 4, width = 2,
                                headlength=4),
               horizontalalignment = 'left', verticalalignment = 'top' 
               )
    
# zoom no período 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financeial crises')
```
