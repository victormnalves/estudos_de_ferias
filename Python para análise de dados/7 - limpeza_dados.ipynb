## Tratamento de ausentes

```python
string_data = pd.Series(['aaaa','bbbb',np.nan,'avocado']) # nan = not a number

string_data
string_data.isnull()

string_data[0] = None
string_data.isnull()

```

#### Métodos para tratamento de NA

dropna() = Filtra rótulos de eixos, assumindo que há valores nulos para aquele rótulo.
<br>
fillna() = Preenche dados ausentes com algum valor ou usando 'ffill' ou 'bfill'.
<br>
isnull() = Verifica se cada valor é nulo ou não.
<br>
notnull() = Verifica se cada valor não é nulo ou não.

### Filtrando dados ausentes

```python
data = pd.Series([1, NA, 3.5, NA, 7])
data.dropna()
```

```python
data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA], [NA, NA, NA], [NA, 6.5, 3.]])
cleaned = data.dropna() #printará as que NÃO TÊM NULOS

data
cleaned

data.dropna(how = 'all') #dropa toda coluna que inteiramente tem nulos
data[4] = NA #toda coluna 3 receberá NAs
data
data.dropna(axis = 1, how = 'all') #dropa do eixo das colunas a coluna que possui somente nulos
```

```python
df = pd.DataFrame(np.random.randn(7,3))
df

df.iloc[:4,1] = NA
df

df.iloc[:2,2] = NA
df

df.dropna()
df.dropna(thresh=2) #queremos a coluna que contém ao menos 2 valores NAs
```

### Preenchendo dados ausentes
É importante resaltar que a função fillna() devolverá um novo objeto.

```python
df.fillna(0) #preenchemos onde há NAs com 0
df.fillna({1:0.5,2:0}) #preenche os NAs e nulos das colunas 2 e 3 com os valores indicados
```

```python
df
df.fillna(0, inplace = True) #faz com que a alteração salve-se no objeto original (salvar inplace)
df
```

```python
df = pd.DataFrame(np.random.randn(6,3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df

df.fillna(method='ffill') #propaga a ultima observação para os valores NA
df.fillna(method='ffill', limit = 2) #limita a propagação até 2 linhas/colunas
```

```python
data = pd.Series([1.,NA,3.5,NA,7])
data.fillna(data.mean()) #preenche com a média
```

#### Argumentos para a função fillna()
value: um valor ou um objeto do tipo dicionário.
<br>
method: Interpolação.
<br>
axis: O eixo a ser preenchido(0: linha, 1: coluna).
<br>
inplace: Retorna a modificação como objeto inplace.
<br>
limit: Número máximo de valores a serem preenchidos.

## Transformação de dados

### Remoção de duplicatas

```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],'k2': [1, 1, 2, 3, 3, 4, 4]})
data

data.duplicated() #verifica se o elemento de uma linha é igual ao seu anterior
data.drop_duplicates() #dropa os valores duplicados

data['v1'] = range(7)
data
data.drop_duplicates(['k1'])
```

### Transformando dados com funções ou mapeamento

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon','corned beef', 'Pastrami'
                             ,'pastrami' , 'Bacon', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data

#criando valores para o mapeamento
meat_to_animal = {'bacon': 'pig','pulled pork': 'pig','pastrami': 'cow',
                  'corned beef': 'cow','honey ham': 'pig','nova lox': 'salmon'}

#transformando os valores da coluna food
lowercased = data['food'].str.lower()
lowercased

#mapeando
data['animal'] = lowercased.map(meat_to_animal) 
data

#mapeando com lambda
data['food'].map(lambda x: meat_to_animal[x.lower()])
```

### Substituindo valores

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data

data.replace(-999, NA)
data.replace([-999,-1000], NA) #substituindo mais que um valor
data.replace([-999,-1000], [NA,0])
```

### Renomeando indices

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['Ohio', 'Colorado', 'New York'],
                    columns=['one', 'two', 'three', 'four'])

#função para mapear em uppercase
transform = lambda x: x[:4].upper()

data.index.map(transform)
```

```python
data.rename(index = str.title, columns = str.upper, inplace=True)
data
```

```python
data.rename(index = {'OHIO': 'INDIANA'}, columns={'three':'peekaboo'})
```

### Discretização e compartimentação (binning)

```python
ages = [20,22,25,27,21,23,37,31,61,45,41,32]
bins = [12,25,35,60,100]

cats = pd.cut(ages,bins) #retornará categorias
cats

cats.codes
cats.categories
pd.value_counts(cats)

#podemos passar labels para os intervalores
group_names = ['Youth','YoungAdult','MiddleAge','Senior']
pd.cut(ages, bins, labels = group_names)
```

```python
data = np.random.randn(20)
pd.cut(data,4,precision = 2) #podemos criar bins implicitamente dividindo os valores e dando uma precisão decimal
pd.value_counts(data)

#separando quartis
data = np.random.randn(1000) #cria os valores normalmente distribuidos
cats = pd.qcut(data, 4) #divide em 4 quartis
cats
pd.value_counts(cats)
print('\n')
pd.qcut(data, [0,0.1,0.5,0.9,1.]) #dividindo em quartis dando como argumento o rótulo
```
